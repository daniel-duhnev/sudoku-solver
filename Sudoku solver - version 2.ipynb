### Version 2: Incorporate 3x3 box check

In this version you will see a 3x3 box check on top of the row and column check before filling in a number as a possibility.
This means there are fewer possibilities in the boxes and so the board can be solved in less steps.
This algorithm can solve slightly harder sudokus than version 1 but is still limited.

```python
import numpy as np
```

```python
# Helper function used in the add all possibilities function below.
# This is added in version 2 of the solver to aid with checking through each square
def add_possibility(board, i, j, k):
    if board[j][k] == 0:
        if board[j][k] != i:
            board[j][k] = []
            board[j][k].append(i)
    elif isinstance(board[j][k],list) and i not in board[j][k]:
        board[j][k].append(i)

# Iterate over each square. This can be tidied up to loop through all the 3x3 squares instead of hardcoding it for each one.
def add_all_possibilities(board):
    for i in range(1,10):
        for j in range(9):
            for k in range(9):
                    if i not in board[j] and i not in np.array(board, dtype=object)[:,k]:
                        if j <= 2:
                            if k <= 2:
                                if i not in np.array(board,dtype=object)[0:3, 0:3]:
                                    add_possibility(board,i,j,k)
                            elif k <=5 and k > 2:
                                if i not in np.array(board,dtype=object)[0:3, 3:6]:
                                    add_possibility(board,i,j,k)
                            elif k <=8 and k > 5:
                                if i not in np.array(board,dtype=object)[0:3, 6:9]:
                                    add_possibility(board,i,j,k)
                        elif j <=5 and j > 2:
                            if k <=2:
                                if i not in np.array(board,dtype=object)[3:6, 0:3]:
                                    add_possibility(board,i,j,k)
                            elif k <=5 and k > 2:
                                if i not in np.array(board,dtype=object)[3:6, 3:6]:
                                    add_possibility(board,i,j,k)
                            elif k <=8 and k > 5:
                                if i not in np.array(board,dtype=object)[3:6, 6:9]:
                                    add_possibility(board,i,j,k)
                        elif j <=8 and j > 5:
                            if k <=2:
                                if i not in np.array(board,dtype=object)[6:9, 0:3]:
                                    add_possibility(board,i,j,k)
                            elif k <=5 and k > 2:
                                if i not in np.array(board,dtype=object)[6:9, 3:6]:
                                    add_possibility(board,i,j,k)
                            elif k <=8 and k > 5:
                                if i not in np.array(board,dtype=object)[6:9, 6:9]:
                                    add_possibility(board,i,j,k)

def fill_box(board):
    for i in range(1,10):
        for j in range(9):
            for k in range(9):
                if isinstance(board[j][k], list) and len(board[j][k]) == 1:
                    board[j][k] = board[j][k][0]

def remove_possibility(board, i, j, k):
    if isinstance(board[j][k],list) and i in board[j][k]:
        board[j][k].remove(i)

def remove_row_column_possibilities(board):
    for i in range(1,10):
        for j in range(9):
            for k in range(9):
                if i in board[j] or i in np.array(board, dtype=object)[:,k]:
                    remove_possibility(board,i,j,k)

def remove_square_possibilities(board):
    for i in range(1,10):
            for j in range(9):
                for k in range(9):
                    if j <= 2:
                        if k <= 2:
                            if i in np.array(board,dtype=object)[0:3, 0:3]:
                                remove_possibility(board,i,j,k)
                        elif k <=5 and k > 2:
                            if i in np.array(board,dtype=object)[0:3, 3:6]:
                                remove_possibility(board,i,j,k)
                        elif k <=8 and k > 5:
                            if i in np.array(board,dtype=object)[0:3, 6:9]:
                                remove_possibility(board,i,j,k)
                    elif j <=5 and j > 2:
                        if k <=2:
                            if i in np.array(board,dtype=object)[3:6, 0:3]:
                                remove_possibility(board,i,j,k)
                        elif k <=5 and k > 2:
                            if i in np.array(board,dtype=object)[3:6, 3:6]:
                                remove_possibility(board,i,j,k)
                        elif k <=8 and k > 5:
                            if i in np.array(board,dtype=object)[3:6, 6:9]:
                                remove_possibility(board,i,j,k)
                    elif j <=8 and j > 5:
                        if k <=2:
                            if i in np.array(board,dtype=object)[6:9, 0:3]:
                                remove_possibility(board,i,j,k)
                        elif k <=5 and k > 2:
                            if i in np.array(board,dtype=object)[6:9, 3:6]:
                                remove_possibility(board,i,j,k)
                        elif k <=8 and k > 5:
                            if i in np.array(board,dtype=object)[6:9, 6:9]:
                                remove_possibility(board,i,j,k)    
```

```python
def sudoku_solver(board):
    # counter variable "n" used to check how many times we need to call the function to solve the puzzle
    global n
    
    add_all_possibilities(board)
    #print(np.array(board, dtype=object))
    
    fill_box(board)
    #print("\n"*2,np.array(board, dtype=object))

    remove_row_column_possibilities(board)
    #print("\n"*2,np.array(board, dtype=object))
    
    remove_square_possibilities(board)
    #print("\n"*2,np.array(board, dtype=object))

    # Produce an array with values True or False depending on whether any of the boxes contain a list of possibilities or not
    # If we still have lists, then we've got work to do so we have to run the function again.

    x = []
    
    for i in np.array(board, dtype=object):
        for j in range(9):
            x.append(isinstance(i[j], list))

    # If there are any True values - i.e. lists that have possible numbers in them, then repeat the loop again.
    if any(i for i in x) is True:
        print("\nThe currect board is:\n",np.array(board, dtype=object))
        # Increase the counter by 1
        n += 1
        sudoku_solver(board)
    # Else if there are no lists left... The puzzle is solved!
    else:
        print("\nSolved in", n+1, "function calls!")
        print("\nThe final board is:\n",np.array(board, dtype=object))
```

```python
# Super easy board
# Same board we started with in version 1 - website: https://www.puzzles-to-print.com/printable-sudokus/sudokus-for-kids/beginner-sudoku-easy.shtml
# Let's see if we can solve this one even more quickly

board_array = [ [5,0,0, 4,6,7, 3,0,9],
                [9,0,3, 8,1,0, 4,2,7],
                [1,7,4, 2,0,3, 0,0,0],

                [2,3,1, 9,7,6, 8,5,4],
                [8,5,7, 1,2,4, 0,9,0],
                [4,9,6, 3,0,8, 1,7,2],

                [0,0,0, 0,8,9, 2,6,0],
                [7,8,2, 6,4,1, 0,0,5],
                [0,1,0, 0,0,0, 7,0,8]
                ]

# Start a counter to see how many times you need to call the function to solve the puzzle
n = 0
sudoku_solver(board_array)
```

In version 1 the above board was solved in 3 function calls. 
So by looking at the squares we limit the amount possible numbers we can put in our temporary lists, and we improve the amount 
of numbers we can remove once we fill in the boxes with only 1 number in their list. These two components are the reason we were able to solve this easy board in 1 less call!

```python
# Medium easy: https://www.puzzles-to-print.com/printable-sudokus/sudokus-for-kids/beginner-sudoku-medium.shtml
board_array = [ [0,5,0, 0,0,3, 7,0,0],
                [8,3,1, 4,2,7, 0,0,5],
                [9,7,4, 0,8,6, 1,2,3],
               
                [3,8,6, 1,4,0, 2,5,7],
                [7,0,5, 0,0,0, 0,0,0],
                [0,4,9, 7,0,2, 8,3,6],
               
                [5,0,0, 0,7,4, 0,9,8],
                [0,9,0, 3,0,1, 0,0,2],
                [2,0,0, 0,9,0, 3,0,1]
                ]

# Start a counter to see how many times you need to call the function to solve the puzzle
n = 0
sudoku_solver(board_array)
```

Similar result here. In this case needing only 4 function calls, compared to 6 in version 1, to solve the puzzle. An improvement indeed!
